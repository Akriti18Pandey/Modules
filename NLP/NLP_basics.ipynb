```python
import nltk
```

```python
nltk.download('punkt_tab')   # to tokenize data
nltk.download('stopwords')   # for common words (is,am,are,and, etc)
nltk.download('wordnet')  #lemmatization dictionary : word to root format
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
stop_words=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
```

```python
import re
```

```python
def clean_data(text):
    text=text.lower()   # converts text to small letters/lowercase
    text=re.sub(r'\s+', ' ',text).strip()  #removes extra spaces
    words=word_tokenize(text)  #split into word list
    words=[w for w in words if w not in stop_words]  # drop unimportant words from sentence
    words=[lemmatizer.lemmatize(w) for w in words] # converts words to their base form
    return "".join(words) #return the clean text
```

```python
review=[
    "The product was amazing and in budget.",
    "Worst mobile ever i have purchased so far , why company design such pathetic phones.",
    "I like this phone model.",
    "Very good product.",
    "Best phone under this budgetand screen size is good !"]
```

```python
cleaned=[clean_data(r) for r in review]

for i in range(len(review)):
    print("Real review:",review[i])
    print("Cleaned review:",cleaned[i])
    print("-" *30)
```
