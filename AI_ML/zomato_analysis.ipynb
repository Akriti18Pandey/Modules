```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("zomato_dataset.csv")
df
```

```python
#No. of rows and columns dataset contains
print("Shape of dataset(rows,columns):", df.shape)
```

```python
# first 5 rows
display(df.head())
```

```python
#last 5 rows
display(df.tail())
```

```python
# to check datatype of all the columns
print(df.dtypes)
```

```python
#No. of missing values present in each columns
print(df.isnull().sum())
```

```python
# No. of unique restaurants present in the dataset
unique_restaurants=df['Restaurant Name'].unique()
print("Number of unique restaurants:",unique_restaurants)
```

```python
# top 10 most common restaurants
print("\nTop 10 restaurants:")
print(df["Restaurant Name"].value_counts().head(10))
```

```python
plt.figure(figsize=(6,4))
df['Restaurant Name'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title("Top 10 Most Common Restaurant Names")
plt.xlabel("Restaurant Name")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# top 5 cuisines serves most frequently
print("Top 5 cuisines:")
print(df['Cuisine'].value_counts().head(5))	
```

```python
plt.figure(figsize=(6,4))
df['Cuisine'].value_counts().head(5).plot(kind='bar', color='lightgreen')
plt.title("Top 5 Most Common Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Restaurants serve multiple cuisines 
multi_cuisine=df[df['Cuisine'].str.contains(',')].shape[0]
print("\nRestaurants serving multiple cuisine:",multi_cuisine)
```

```python
# Average cost for two people
avg_cost= df['Cost for Two'].mean()
print("Average cost for two people:",avg_cost)
```

```python
# location with highest Average cost for two 
location=df.groupby('Location')['Cost for Two']
print("\n location with highest Average cost for two:")
print(location.head(1))
```

```python
#correlation between cost and rating
correlation= df['Cost for Two'].corr(df['Rating'])
print("\nCorrelation between cost and rating:",correlation)
```

```python
sns.set(style="whitegrid")
```

```python
#Scatter plot: Cost vs Rating
plt.figure(figsize=(6,3))
sns.scatterplot(x='Cost for Two', y='Rating', data=df, color='purple')
plt.title("Cost vs Rating")
plt.xlabel("Cost for Two")
plt.ylabel("Rating")
plt.show()
```

```python
# restaurant's maximum and minimum cost for two
max_cost=df[df['Cost for Two']== df['Cost for Two'].max()]
min_cost=df[df['Cost for Two']== df['Cost for Two'].min()]
```

```python
print("\nRestaurant's with maximum cost for two:")
display(max_cost[['Restaurant Name','Location','Cost for Two']])	
```

```python
print("\nRestaurant's with minimum cost for two:")
display(min_cost[['Restaurant Name','Location','Cost for Two']])	
```
