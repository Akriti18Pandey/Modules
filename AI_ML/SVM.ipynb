```python
import pandas as pd
```

```python
df=pd.read_csv("cat_dog_svm_dataset.csv")
df
```

```python
# map= convert string into binary
df['Animal']=df['Animal'].map({'Cat':0,'Dog':1})
df
```

```python
x=df[['WeightKg','HeightCm','EarType','TailLengthCm']]
y=df['Animal']
```

```python
#split data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVC

model=SVC(kernel='linear')  # separate the classes into a straight line
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
new=[[5,20,0,5]]
pred=model.predict(new)

print("Animal is :","Cat" if pred==0 else "Dog")
```
