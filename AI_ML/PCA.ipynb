```python
import pandas as pd
```

```python
df=pd.read_csv("student_pca_dataset.csv")
df
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
model=pca.fit_transform(df)
```

```python
pca_df=pd.DataFrame(model, columns=['A1','A2'])
pca_df
```

```python
pca.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(pca_df['A1'],pca_df['A2'],color="green")
plt.xlabel("Principal Comp 1")
plt.ylabel("Principal Comp 2")
plt.title("Student Performance")
plt.show()
```
