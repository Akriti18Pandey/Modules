```python
import pandas as pd
import numpy as np
```

```python
Spending=[45, 60, 40, 55, 70, 65,50, 45, 90, 85, 60, 55, 50, 45, 70, 75, 65, 50, 55, 60, 80, 85, 75, 60, 55, 45, 70, 65, 60, 55]

```

```python
df=pd.DataFrame(Spending, columns = ["Spending"])
df
```

```python
from scipy import stats
```

```python
print("Mean:",np.mean(Spending))
print("Median:",np.median(Spending))
```

```python
print("Mode:",stats.mode(Spending))
```

```python
print("Range:",max(Spending)-min(Spending))
```

```python
print("Standard Deviation:",np.std(Spending))
```

```python
bins=[40,50,60,70,80,90,100]
freq =pd.cut(df["Spending"],bins=bins).value_counts().sort_index()
freq
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(Spending, bins=bins, edgecolor="black")
plt.title("Histogram of Spending")
plt.xlabel("Spending")
plt.ylabel("Frequency")
plt.show()
```

```python
Q1=df["Spending"].quantile(0.25)
Q3=df["Spending"].quantile(0.75)
iqr=Q3-Q1
```

```python
# HOW MUCH MIDDLE DATA SPREAD FROM MEAN VALUE
iqr
```

```python
#range=x,y
x=Q1 - 1.5 * iqr
y=Q3 + 1.5 * iqr
```

```python
Outliers=df[(df["Spending"]<x) | (df["Spending"]>y)]
```

```python
heights = [160, 165, 170, 168, 172, 169, 171, 167, 166, 173,
           175, 169, 168, 170, 172, 174, 165, 167, 168, 171]
```

```python
df=pd.DataFrame(heights)
df
```

```python
mean=np.mean(df)
print("Mean Height:",mean)
```

```python
SD=np.std(df)
CV=(SD / mean)*100
```

```python
print("Standard Deviation:",round(SD,2))
print("Covariance:",round(CV,2),"%")
```
