```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("customers_seg.csv")
data.head()
```

```python
X = data[['Age','Income','Spending_Score']]
```

```python
# Q-1-Apply clustering

kmeans = KMeans(n_clusters=5, random_state=42)
data['Cluster'] = kmeans.fit_predict(X)
data.head()
```

```python
# Q-2-Visualize clusters

plt.figure(figsize=(8,6))

plt.scatter(data['Income'], data['Spending_Score'],
            c=data['Cluster'], cmap='viridis')

plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation Clusters")

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(data['Age'], data['Income'],
            c=data['Cluster'], cmap='rainbow')

plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Customer Groups based on Age & Income")

plt.show()
```

```python
# Q-3-INTERPRET Groups

cluster_summary = data.groupby('Cluster').mean()

print(cluster_summary)
```
