```python
import pandas as pd
```

```python
df=pd.read_csv("customer_segmentation.csv")
df
```

```python
from sklearn.cluster import KMeans
```

```python
KM=KMeans(n_clusters=6, random_state=42)
df["Cluster"]=KM.fit_predict(df)
```

```python
print(df.head())
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['AnnualIncome'],df['SpendingScore'],c=df['Cluster'])
plt.xlabel("AnnualIncome")
plt.ylabel("Spending")
plt.title("Customer Segmentation")
plt.show()
```

```python
## CLUSTERING ON THE BASIS OF STUDENTS MARKS
```

```python
df=pd.read_csv("student_marks_clustering.csv")
df
```

```python
x=df[["Total"]]
```

```python
KMs=KMeans(n_clusters=3, random_state=42)
df['Cluster']=KMs.fit_predict(x)
```

```python
plt.scatter(df['StudentID'],df['Total'],c=df['Cluster'])
plt.xlabel("Student ID")
plt.ylabel("Total Marks")
plt.title("Students Performance Cluster")
plt.show()
```
