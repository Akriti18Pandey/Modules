```python
import speech_recognition as sr
```

```python
recognizer=sr.Recognizer()   # It is controller which listen to audio and convert to text

with sr.Microphone() as source:   # It open the microphone as the audio source
    recognizer.adjust_for_ambient_noise(source, duration=1)   # To measure the background noise
    print("Speak now:")
    audio=recognizer.listen(source)  # Records the voice untill we stop speaking

try:
    text=recognizer.recognize_google(audio,language="en-US")  # it send the recorded audio to google's Speech Recognition
    print("English Text:",text)

    text=recognizer.recognize_google(audio,language="hi-IN")
    print("Hindi Text:",text)
except sr.UnknownValueError:        
    print("Could not understand")
```
