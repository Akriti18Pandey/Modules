```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("tata_motors.csv")
df
```

```python
prices=df['Close'].values.reshape(-1,1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()  # it scales values into range (0 to 1)
price_scaled=scaler.fit_transform(prices)  #it learns the min/max from our data and transform accordingly
```

```python
#creating sequences last 5 day > next day
x=[]
y=[]

for i in range(len(price_scaled)-5):
    x.append(price_scaled[i:i+5])  #says previousprices (features)
    y.append(price_scaled[i+5])  # the next value(target)

x=np.array(x)
y=np.array(y)
```

```python
# Build RNN Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input,LSTM,Dense
```

```python
model=Sequential()

model.add(Input(shape=(5,1)))
model.add(LSTM(50))  #50 memory cells to learn patterns
model.add(Dense(1))    #output one value(predicted)

model.compile(optimizer='adam',loss='mse')    # for regression
```

```python
#train model
model.fit(x,y, epochs=10)
```

```python
# predict Next Day Price

five_day=price_scaled[-5:]
five_day=five_day.reshape(1,5,1)

prediction=model.predict(five_day)
predicted_price=scaler.inverse_transform(prediction)

print("Prediction of the next day close price is:",predicted_price)
```
