```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Input
```

```python
n=1000

data={"Age":np.random.randint(15,50,n),  #age between 15 to 50
      "Genre":np.random.randint(0,5,n),  #movies category (action=0, drama=1,fiction=2, comedy=3, horror=4)
      "Rating":np.random.randint(1,6,n), # rating from 1 to 5 star
      "WatchTime":np.random.randint(10,180,n)  # watch time 10min to 180 min)
     }
```

```python
df=pd.DataFrame(data)
df
```

```python
#rule-based
df['Like']=(
    (df['Rating'] >3)&
    (df['WatchTime'] >60 )).astype(int)
```

```python
df
```

```python
#split
x=df.drop("Like",axis=1)
y=df["Like"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scale=StandardScaler()

x_train=scale.fit_transform(x_train)    # it calculates the mean and SD of each features in training set
x_test=scale.transform(x_test)
```

```python
model=Sequential([
    Input(shape=(x_train.shape[1],)),
    Dense(12,activation='relu'),
    Dense(8,activation='relu'),
    Dense(1,activation='sigmoid')])
```

```python
#compile
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy'])
```

```python
# train model
model.fit(x_train,y_train,epochs=20,batch_size=16)
```

```python
# evaluation
loss,accuracy=model.evaluate(x_test,y_test)
print("Accuracy:",accuracy)
```

```python
#test new data
new_data=np.array([[26,3,4.5,140]])  # age,genre,rating,watchtime
```

```python
new_data_scale=scale.transform(new_data)
```

```python
predict=model.predict(new_data_scale)
```

```python
final=(predict > 0.5).astype(int)

print("Prediction 0=Dislike 1=Like:",final[0][0])
```
