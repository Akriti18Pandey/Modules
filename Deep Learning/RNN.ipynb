```python
#English- Hindi model

model="Helsinki-NLP/opus-mt-en-hi"
```

```python
tokenizer=MarianTokenizer.from_pretrained(model)
models=MarianMTModel.from_pretrained(model)
```

```python
text=["A sleeping lion is awakened by a tiny mouse, but rather than eating it, the lion generously spares its life. Later, when the lion is caught in a hunter's net, the same tiny mouse proves her worth by gnawing through the ropes to set him free, showing that even the smallest friend can be of great help."]
```

```python
tokens=tokenizer(text,return_tensors="pt",padding="max_length")
translated=models.generate(**tokens)
```

```python
output=tokenizer.decode(translated[0],skip_special_tokens=True)
print("Translated text is :",output)
```
